```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
%matplotlib inline
```

```python
df_Q1 = pd.read_csv('HealthyRideRentals2019-Q1.csv')
df_Q2 = pd.read_csv('HealthyRideRentals2019-Q2.csv')
df_Q3 = pd.read_csv('HealthyRideRentals2019-Q3.csv')
frames = [df_Q1, df_Q2, df_Q3]



df = pd.concat(frames) # create raw dataframe
day_res_df = pd.DataFrame(columns = ['date', 'station', 'fromCNT', 'toCNT', 'rebalCNT']) 

# 1/1/19 starting date
starting_date = [1, 1, 19] 
# get all station IDs
unique_ids = df['From station id'].unique()

# iterate through dates
while not(starting_date[0] == 10 and starting_date[1] == 3): # end date is 10/2/19
    # get all the rows of the specific date
    day_df = df.loc[df['Starttime'].str.split().str[0] == '/'.join([str(elem) for elem in starting_date])]
    # if specific date has transactions
    if len(day_df)>0:
        # iterate through each station
        for station_id in unique_ids:
            fromCNT = (day_df['From station id'].values == station_id).sum() # sum fromCNT
            toCNT = (day_df['To station id'].values == station_id).sum() # sum toCNT
            rebalCNT = abs(fromCNT - toCNT) # find approx rebalCNT
            # add them to the results dataframe
            day_res_df = day_res_df.append({'date' : '/'.join([str(elem) for elem in starting_date]), 'station' : station_id, 'fromCNT' : fromCNT, 'toCNT' : toCNT, 'rebalCNT' : rebalCNT, }, 
                ignore_index = True)
    
    # increase day...
    starting_date[1] += 1
    if starting_date[1] > 31: # end of month
        starting_date[0] += 1 # increase month
        starting_date[1] = 1 # reset day to 1
day_res_df = day_res_df.sort_values(by=['date', 'station'])
day_res_df.head(20)

```

```python

month_res_df = pd.DataFrame(columns = ['date', 'station', 'fromCNT', 'toCNT', 'rebalCNT']) 
month_df = pd.DataFrame()
# 1/1/19 starting date
starting_date = [1, 1, 19] 
# get all station IDs
unique_ids = df['From station id'].unique()

# iterate through dates
while not(starting_date[0] == 10 and starting_date[1] == 3): # end date is 10/2/19
    # get all the rows of the specific date
    day_df = df.loc[df['Starttime'].str.split().str[0] == '/'.join([str(elem) for elem in starting_date])]
    # append specific day dataframe to month dataframe
    month_df = month_df.append(day_df)
    #print(month_df)
    if starting_date[1] > 31: # end of month
        # if month has transactions
        if len(month_df)>0:
            # iterate through each station
            for station_id in unique_ids:
                fromCNT = (month_df['From station id'].values == station_id).sum() # sum fromCNT
                toCNT = (month_df['To station id'].values == station_id).sum() # sum toCNT
                rebalCNT = abs(fromCNT - toCNT) # find approx rebalCNT
                # add them to the results dataframe
                month_res_df = month_res_df.append({'date' : starting_date[0], 'station' : station_id, 'fromCNT' : fromCNT, 'toCNT' : toCNT, 'rebalCNT' : rebalCNT, }, 
                    ignore_index = True)
        
        starting_date[0] += 1 # increase month
        starting_date[1] = 1 # reset day to 1
    
    
    else:
        # increase day...
        starting_date[1] += 1
month_res_df = month_res_df.sort_values(by=['date', 'station'])
month_res_df.head(20)
```

```python


filter_month = 4
filter_stationID = 1046
```

```python

filter_month_df = month_res_df.loc[month_res_df['date'] == filter_month]
ordered_fromCNT_df = filter_month_df.nlargest(25, 'fromCNT')
ordered_fromCNT_df.plot.bar(x="station", y="fromCNT", title='Stations With Highest fromCNT of Given Month')
```

```python

all_transactions_of_month_df = df.loc[df['Stoptime'].str.split().str[0].str.split('/').str[0] == str(filter_month)]
daily_distribution_df = pd.DataFrame()
for hour in range(24):
    nth_hour_df = all_transactions_of_month_df.loc[all_transactions_of_month_df['Stoptime'].str.split().str[1].str.split(':').str[0] == str(hour) ]
    daily_distribution_df = daily_distribution_df.append( {'hour' : hour, 'fromCNT' : (nth_hour_df['From station id']).sum()}, ignore_index=True )

daily_distribution_df.plot.line(x='hour', y='fromCNT', title="Hourly fromCNT of Given Month")
```

```python

all_transactions_of_month_df = df.loc[df['Stoptime'].str.split().str[0].str.split('/').str[0] == str(filter_month)]
unique_bikes = all_transactions_of_month_df['Bikeid'].unique()
popular_bikes_df = pd.DataFrame()

for bike in unique_bikes:
    bike_sum = (all_transactions_of_month_df['Bikeid'].values == bike).sum()
    popular_bikes_df = popular_bikes_df.append({ 'bikeid' : bike, 'Number of Times Bike was Rented' : bike_sum}, ignore_index=True)

ordered_bikes = popular_bikes_df.nlargest(25, 'Number of Times Bike was Rented')
ordered_bikes.plot.bar(x="bikeid", y="Number of Times Bike was Rented", title="Most Rented Bikes of Given Month")
```

```python
# 6-dimensional space data frame
six_df = pd.DataFrame()
for station_id in unique_ids:
    tuples = []
    for month in range(7, 10):
        filter_month_df = month_res_df.loc[(month_res_df['date'] == month) & (month_res_df['station'] == station_id)]
        total_fromCNT = filter_month_df['fromCNT'].sum()
        total_rebalCNT = filter_month_df['rebalCNT'].sum()
        tuples.append(tuple((total_fromCNT, total_rebalCNT)))
    month7_fromCNT, month7_rebalCNT = tuples[0]
    month8_fromCNT, month8_rebalCNT = tuples[1]
    month9_fromCNT, month9_rebalCNT = tuples[2]
    six_df = six_df.append({ 'month7_fromCNT' : month7_fromCNT, 'month7_rebalCNT' : month7_rebalCNT, 'month8_fromCNT' : month8_fromCNT, 'month8_rebalCNT' : month8_rebalCNT, 'month9_fromCNT' : month9_fromCNT, 'month9_rebalCNT' : month9_rebalCNT,}, ignore_index=True )
#'station' : station_id, 

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN

kmeans2 = KMeans(n_clusters=2).fit(six_df)
kmeans3 = KMeans(n_clusters=3).fit(six_df)
kmeans6 = KMeans(n_clusters=6).fit(six_df)

db1 = DBSCAN(eps=50, min_samples=2).fit(six_df)   
db2 = DBSCAN(eps=100, min_samples=2).fit(six_df) 
db3 = DBSCAN(eps=150, min_samples=5).fit(six_df) 

```

```python

from collections import Counter
from collections import OrderedDict


dt = dict(Counter(kmeans2.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='KMeans k = 2')


dt = dict(Counter(kmeans3.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='KMeans k = 3')


dt = dict(Counter(kmeans6.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='KMeans k = 6')


dt = dict(Counter(db1.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    if key == -1: # IGNORE NOISY POINTS
        continue
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='DBSCAN eps=50, min_samples=2')


dt = dict(Counter(db2.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    if key == -1: # IGNORE NOISY POINTS
        continue
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='DBSCAN eps=100, min_samples=2')


dt = dict(Counter(db3.labels_))
sorted_dict = {k: v for k, v in sorted(dt.items(), key=lambda item: item[1], reverse=True)}
sorted_values = []
sorted_keys = []
for key, val in sorted_dict.items():
    if key == -1: # IGNORE NOISY POINTS
        continue
    sorted_keys.append(key)
    sorted_values.append(val)
data = {
    'Cluster ID' : sorted_keys,
    'Number of Stations' : sorted_values,
}
bar_df = pd.DataFrame(data)
bar_df.plot.bar(x="Cluster ID", y="Number of Stations", title='DBSCAN eps=150, min_samples=5')

```

```python

from sklearn.neighbors import NearestNeighbors
import numpy as np

# ELBOW TEST
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(six_df)
    distortions.append(kmeanModel.inertia_)
plt.figure(figsize=(8,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('KMeans Elbow Method')
plt.show()

# KNEE TEST
NN = NearestNeighbors(n_neighbors=5)
NN.fit(six_df)
dist, indices = NN.kneighbors(six_df)
dist = np.sort(dist, axis=0)[:, 1]
plt.plot(dist)
plt.xlabel('Sorted Points')
plt.ylabel('Distance')
plt.title('DBSCAN Knee Method')
plt.show()

```

My choices for k were predicated on the Elbow Method shown above. I chose 2, 3, and 6, as those were ponts along the Elbow curve which looked most interesting to me, as 2 and 3 are closest to the "elbow", and 6 was more of an expiremental/outlier choice. 
A) Looking at the KMeans Elbow Method plotted above, it would appear that k=3 is the 'elbow' of the curve, highlighting that k=3 is most likely closest to the optimal k value, as after k=3, the k+1 cluster would not give a better model of the data with consideration for the risk of overfitting.  
B) For this data set, the results are inconclusive as to which algorithm is preferred. Looking at the DBSCAN Knee plot and KMeans Eblow plot, it is reasonble to assume that the "best" DBSCAN implemented is the one with an eps=150 and a min_samples=5, and the "best" KMeans is where k=3. However, if we assuming that the data set will continue to grow, and that there is a possibility that more dimensions will be added in the future (ie, month10_fromCNT, month10_rebalCNT, etc), than KMeans would probably be the preferred algorithm for three reasons: 1) it is simply faster and more efficient in terms of scalability, 2) it can handle the high-dimensional data better than DBSCAN (according to class lecture slides), 3) there is no guarantee that we are dealing with data of consistent density, and DBSCAN suffers at handling different densities (according to class lecture slides).
